```python
import pandas as pd
```

```python
df = pd.read_csv("Car_Data_v3.csv")
```

```python
df.head(2)
```

```python
df.drop(["name","torque"],axis = 1, inplace= True)
```

```python
df.head(2)
```

```python
df.owner.unique()
```

```python
df = df[df.owner != "Fourth & Above Owner" ]
df.owner.unique()
```

```python
df.isnull().sum()
```

```python
df = df[~df.mileage.isnull()]
df
```

```python
df.isnull().sum()
```

```python
df.owner.unique()
```

```python
df.seller_type.unique()
```

```python
df["fuel"] = df.fuel.replace("LPG","CNG")
df.fuel.unique()
```

```python
l = []
for i in df.year:
    a = 2023 - i
    l.append(a)
```

```python
df["year"] = l
df.head()
```

```python
l1 = []
for i in df.mileage:
    b = float(i.split(" ")[0])
    l1.append(b)
```

```python
df["mileage_kmpl"] = l1
```

```python
l2 = []
for i in df.engine:
    c = float(i.split(" ")[0])
    l2.append(c)
```

```python
df["engine_CC"] = l2
```

```python
df.head()
```

```python
df.drop(["engine","mileage"],axis = 1,inplace = True)
```

```python
df.head()
```

```python
l3 = []
for i in df.max_power:
    d = pd.to_numeric(i.split(" ")[0])
    l3.append(d)
```

```python
df["max_power_bhp"] = l3
```

```python
df.drop("max_power",axis = 1, inplace = True )
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[~df.max_power_bhp.isnull()]
df.head()
```

```python
df.isnull().sum()
```

```python
df1 = pd.get_dummies(data = df ,columns = ["fuel","seller_type", "transmission", "owner"],drop_first = True)
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
l = ["year","selling_price","km_driven","seats","mileage_kmpl","engine_CC"]
for i in l:
    df1[i] = scaler.fit_transform(df1[[i]])
```

```python
df1.head()
```

```python
X = df1.drop("selling_price", axis = 1)
y = df1.selling_price
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test ,y_test)
```

```python
prediction = model.predict(X_test)
prediction
```
