```python
import pandas as pd
```

```python
df = pd.read_csv("Car_Data_v3.csv")
```

```python
df.info()
```

```python
df.drop("name",axis=1,inplace=True)
df.head()
```

```python
df['Present_year'] = 2023
df.head()
```

```python
df['year'] = df['Present_year'] - df['year']
```

```python
df.head()
```

```python
df.drop("Present_year",axis=1,inplace=True)
df.head()
```

```python
df['fuel'].unique()
```

```python
df.fuel.replace("LPG","CNG",inplace = True)
df['fuel'].unique()
```

```python
df['transmission'].unique()
```

```python
df[~(df['owner']=='Fourth & Above Owner')]
```

```python
df = pd.get_dummies(data = df, columns=['fuel','transmission','owner'],drop_first =True)
df.head()
```

```python
df["mileage"].unique()
```

```python
l = []
for i in df["mileage"]:
    try:
        i.split(" ")
        a = i.split(" ")[0]
        l.append(a)
    except:
        l.append(i)
```
